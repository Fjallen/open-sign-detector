```python
import cv2
```

```python
vidcap = cv2.VideoCapture('none.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("./none/none%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```

```python
vidcap = cv2.VideoCapture('leftsign.mp4')
success,image = vidcap.read()
count = 0
while (count <= 2):
  cv2.imwrite("./left/left%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
